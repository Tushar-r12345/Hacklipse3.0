```python
import json
from sklearn.preprocessing import LabelEncoder

# Create an example label encoder object
# le = LabelEncoder()
# le.fit(["red", "green", "blue", "red"])

# Get the inverse mapping
inverse_mapping_diagnose = le1.inverse_transform(range(len(le1.classes_)))

# Create the dictionary of column name to inverse mapping
column_mapping = {"diagnose": {k: v for k, v in zip(range(len(le1.classes_)), inverse_mapping_diagnose)}}

# Convert the dictionary to a JSON string
json_str = json.dumps(column_mapping)

# Write the JSON string to a file
with open("inverse_mapping_diagnose.json", "w") as f:
    f.write(json_str)
```

# model training begins from here

```python
X=df7.drop(['Diagnosis','Treatment'],axis='columns')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# split data into train and test sets
X_train, X_test, y_train_diagnosis, y_test_diagnosis, y_train_treatment, y_test_treatment = train_test_split(X, y1_encoded, y2_encoded, test_size=0.2, random_state=42)

# train random forest classifier for diagnosis
rfc_diagnosis = RandomForestClassifier(n_estimators=100, random_state=42)
rfc_diagnosis.fit(X_train, y_train_diagnosis)

# train random forest classifier for treatment
rfc_treatment = RandomForestClassifier(n_estimators=100, random_state=42)
rfc_treatment.fit(X_train, y_train_treatment)
```

```python
rfc_treatment.score(X_test,y_test_treatment)
```

```python
rfc_diagnosis.score(X_test,y_test_diagnosis)
```

```python
def predict_disease(Previous_illnesses,Cardiac_CT,Age, Chest_pain, Shortness_of_breath, Systolic, Diastolic,
       Heart_rate, Cholesterol_level, Diabetes, Hypertension, Smoking, Obesity, Gender):
    
    # Initialize x array with zeros of length equal to number of columns in X
    x = np.zeros(len(X.columns))
    
    # Find the index of Cardiac_CT and Previous_illnesses in the X.columns array
    loc_index1 = np.where(X.columns == Previous_illnesses)[0][0]
    loc_index2 = np.where(X.columns == Cardiac_CT)[0][0]
    
    # Set the corresponding feature values in the x array
    x[2] = Age
    x[3] = Chest_pain
    x[4] = Shortness_of_breath
    x[5] = Systolic
    x[6] = Diastolic
    x[7] = Heart_rate
    x[8] = Cholesterol_level
    x[9] = Diabetes
    x[10] = Hypertension
    x[11] = Smoking
    x[12] = Obesity
    x[13] = Gender
    
    # Set the values of the Cardiac_CT and Previous_illnesses features to 1
    if loc_index1>0:
        x[loc_index1]=1
    if loc_index2>0:
        x[loc_index2]=1
    
    # Use the trained random forest classifier model to predict the diagnosis
    diagnosis = rfc_diagnosis.predict([x])[0]
    treatment = rfc_treatment.predict([x])[0]
    
    
    # Example encoded result
    encoded_result1 = diagnosis
    encoded_result2 = treatment

    # Look up the original value
    original_value1 = diagnose_mapping[str(encoded_result1)]
    original_value2 = treatment_mapping[str(encoded_result2)]
    
    return original_value1,original_value2
 
```

```python
predict_disease('no serious illness','25% - 45% dilation of the infrarenal aorta',65,1,1,130,80,100,220,0,1,1,0,1)
```
